```python
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
```

```python
iris = datasets.load_iris()
X = iris.data # take all 4 features
y = iris.target # take all 3 species of flower
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

# Gini Impurity Measure with Tree Depth of 2 and 5

```python
gini = DecisionTreeClassifier(criterion='gini',max_depth=2,random_state=1)
gini.fit(X_train, y_train)

print('Gini Depth 2 Mean Accuracy:', gini.score(X_test, y_test))

export_graphviz(gini, out_file='gini5.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('gini5.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
gini = DecisionTreeClassifier(criterion='gini',max_depth=5, random_state=1)
gini.fit(X_train, y_train)

print('Gini Depth 5 Mean Accuracy:', gini.score(X_test, y_test))

export_graphviz(gini, out_file='gini8.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('gini8.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

# Entropy Impurity Measure with Tree Depth of 2 and 5

```python
entropy = DecisionTreeClassifier(criterion='entropy',max_depth=2,random_state=1)
entropy.fit(X_train, y_train)

print('Entropy Depth 2 Mean Accuracy:', entropy.score(X_test, y_test))

export_graphviz(entropy, out_file='entropy5dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('entropy5dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
entropy = DecisionTreeClassifier(criterion='entropy',max_depth=5, random_state=1)
entropy.fit(X_train, y_train)

print('Entropy Depth 5 Mean Accuracy:', entropy.score(X_test, y_test))

export_graphviz(entropy, out_file='entropy8.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('entropy8.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

# Dicussion

| Impurity Measure | Depth | Accuracy |
| --- | --- | --- |
| Gini | 2 | 0.9166|
| Entropy | 2 | 0.9166|
| Gini | 5 | 0.9333|
| Entropy | 5 | 0.9333|

In this assignment, I was asked to use the 2 differnet impurity measures on the Iris dataset using 4 features and all 3 species. Gini measurement is the probability of classifyning a random sample incorrectly by picking a label in the branch. And entropy is the measurement of information, the tree will use the information to make a split. 

I noticed that the accuracy from both impurity measures are the same at the same tree depth. That is, at depth 2, gini and entropy measure have accuracy of 0.9333 and at depth 5, gini and entropy has accuracy of 0.9467. From my observation above, I can conclude that gini and entropy impurity measures are negligible in difference and are pretty much identical, for this dataset at least.

Upon further research by Laura Elena Raileanu and Kilian Stoffel, they compared both types of measurement in "Theoretical comparison between the gini index and information gain criteria", and it is noted that it only matters which of the two measures you choose in 2% of the cases. Therefore, in most cases, the results would be the same with either one of the impurity measures.
